```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

## BAR CHART

```python
fruits = ['Apples', 'Bananas','Cherries','Dates']
sales = [400,350,300,450]


plt.bar(fruits, sales, color='g', width=0.3)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
barWidth = 0.25
fig = plt.subplots(figsize=(12,8))

IT = [12, 30,1,8,22]
ECE = [28,6,16,5,10]
CSE = [29,3,24,4,5]


br1= np.arange(len(IT))
br2 = [x+barWidth for x in br1]
br3 = [x+barWidth for x in br2]
plt.bar(br1,IT, color='r',width=barWidth, edgecolor='grey', label='IT')
plt.bar(br2,ECE, color='g',width=barWidth,edgecolor='grey', label='ECE')
plt.bar(br3,CSE, color='b',width=barWidth, edgecolor='grey', label='CSE')

plt.xlabel('Branch', fontweight='bold', fontsize=15)
plt.ylabel('Students Passed', fontweight='bold', fontsize=15)

plt.xticks([r + barWidth for r in range(len(IT))],
           ['2015','2016','2017','2018','2019'])

plt.legend() 

plt.show()
```

## HISTOGRAM

```python
data = np.random.normal(10.0,3,500)
sns.displot(data, kde= True, bins=10, color='black')
```

## PIE CHART

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [23,17,35,29,12,41]

def func(pct, allvalues):
    absolute = int(pct/100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

#creating plot
fig = plt.figure(figsize=(10,7))
plt.pie(data, labels=cars, autopct = lambda pct:func(pct, data))

#show plot
plt.show()
```

## SCATTER PLOT

```python
x = np.array([12,45,7,32,89,54,23,67,14,91])
y = np.array([99,31,72,56,19,88,43,61,35,77])

plt.scatter(x,y, label = 'Data point')
plt.title('Scatter plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
x1 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y1 = np.array([55,56,72,60,66,62,72,74,80,88])

x2 = np.array([150, 155, 160, 165, 170, 175, 180, 185, 190, 195])
y2 = np.array([52,51,76,68,61,57,78,70,74,81])

plt.scatter(x1, y1, color='blue', label='Group 1')
plt.scatter(x2, y2, color='red', label='Group 2')

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Comparison of Height vs Weight between two groups')

plt.legend(loc='lower right')
plt.show()
```

## Box Plot

```python
# Sample Data
np.random.seed(10)
data = np.random.normal(loc=50, scale=15, size=200)  
# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)

plt.title('Box Plot of Sample Data', fontsize=14)
plt.xlabel('Sample Data', fontsize=12)

plt.show()
```
