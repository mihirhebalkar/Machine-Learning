````python
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=200)

# Create a figure
plt.figure(figsize=(12, 10))

# Histogram
plt.subplot(2, 2, 1)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Boxplot
plt.subplot(2, 2, 2)
sns.boxplot(x=data, color='lightcoral')
plt.title("Boxplot")

# Violinplot
plt.subplot(2, 2, 3)
sns.violinplot(x=data, color='lightgreen')
plt.title("Violin Plot")

# Histogram with KDE
plt.subplot(2, 2, 4)
sns.histplot(data, bins=20, kde=True, color='purple')
plt.title("Histogram with KDE")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
````

# **EDA - Central Tendency, Range, and Plots**

### **1. Central Tendency**

- **Mean** (Arithmetic Average):
  - Sum of data points by number of observations.

- **Median** (Middle Value of Sorted Data):
  - If \(n\) is odd: Middle element of the sorted data.
  - If \(n\) is even: Average of the two middle elements.

- **Mode** (Most Frequent Value):
  - The value that occurs most frequently in the dataset.

### **2. Range**

- **Range** (Difference Between Max and Min Values):
  \[
    {Range} = max(x) - min(x)
  \]
  Where:
  - \(max(x)\) = maximum value in the dataset
  - \(min(x)\) = minimum value in the dataset

### **3. Plots**

- **Histogram**:
  - Visualizes the frequency distribution of a numerical variable.
  - Helps identify skewness, peaks, and spread of data.
  
  - Code Example (Python):
    ```python
    import matplotlib.pyplot as plt
    plt.hist(data, bins=20)
    plt.title('Histogram')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    ```

- **Box Plot**:
  - Shows the spread of the data using quartiles (min, Q1, median, Q3, max).
  - Identifies outliers (values outside 1.5 times the IQR from the quartiles).
  
  - Code Example (Python):
    ```python
    import seaborn as sns
    sns.boxplot(data=data)
    plt.title('Box Plot')
    plt.show()
    ```

- **KDE Plot** (Kernel Density Estimate):
  - Smooths the data to estimate the probability density function of a continuous variable.
  
  - Code Example (Python):
    ```python
    sns.kdeplot(data, shade=True)
    plt.title('KDE Plot')
    plt.show()
    ```

### **4. Summary**

- **Central Tendency**: Measures the "center" of the data (Mean, Median, Mode).
- **Range**: Shows the spread of the data from the minimum to the maximum.
- **Plots**: Visualize the distribution and patterns in the data (Histogram, Box Plot, KDE Plot).
