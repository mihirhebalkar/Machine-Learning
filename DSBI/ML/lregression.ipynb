```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
url = "http://lib.stat.cmu.edu/datasets/boston"

raw_df = pd.read_csv(url, sep="\s+", skiprows=22, header=None)
```

```python
X = np.hstack([raw_df.values[::2,:], raw_df.values[1::2,:2]])
y =raw_df.values[1::2,2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 1)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print('Coefficients : ', regressor.coef_)

print('Variance score : {}'.format(regressor.score(X_test, y_test)))
```

```python
plt.style.use('fivethirtyeight')

plt.scatter(regressor.predict(X_train), regressor.predict(X_train)-y_train, color='g', s=10, label='Train Data')
plt.scatter(regressor.predict(X_test), regressor.predict(X_test)-y_test, color='b', s=10, label='Test Data')
plt.hlines(y=0, xmin=0, xmax=50, linewidth=2)

plt.legend(loc='upper right')

plt.title('Residual Errors')

plt.show()
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/nguyen-toan/ISLR/refs/heads/master/dataset/Advertising.csv')
```

```python
X = dataset[['TV', 'Radio','Newspaper']]
y = dataset['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
mlr  = LinearRegression()

mlr.fit(X_train, y_train)
```

```python
print("Coefficients : ", mlr.coef_ )

print("Intercept : ", mlr.intercept_)
```

```python
y_pred = mlr.predict(X_test)

print("Prediction for test set : {} ".format(y_pred))
```

```python
mlr_diff = pd.DataFrame({'Actial value' : y_test, 'Predicted value' : y_pred})
mlr_diff.head()
```

```python
from sklearn import metrics
meanAbsErr = metrics.mean_absolute_error(y_test, y_pred)
meanSqErr = metrics.mean_squared_error(y_test, y_pred)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("Mean Absolute Error : ", meanAbsErr)
print("Mean Squared Error : ", meanSqErr)
print("Root Mean Squared Error : ", rootMeanSqErr)
```
