```python
# Fill missing values for numeric columns
df.fillna(df.median(numeric_only=True), inplace=True)

# Encode Labels (Convert attack names to numbers)
df['Label'], unique_labels = pd.factorize(df['Label']) 

# Print the mapping of attacks to numbers
label_mapping = {index: attack_name for index, attack_name in enumerate(unique_labels)}
print("ðŸ”¹ Attack Label Mapping:")
for label_num, attack_name in label_mapping.items():
    print(f"{label_num}: {attack_name}")
```

```python
attack_mapping = {
    0: "Benign",
    1: "Network Attack",
    2: "Network Attack",
    3: "Botnet Attack",
    4: "Infiltration Attack",
    5: "Web Attack",
    6: "Web Attack",
    7: "Web Attack",
    8: "Brute Force Attack",
    9: "Brute Force Attack",
    10: "DoS Attack",
    11: "DoS Attack",
    12: "DoS Attack",
    13: "DoS Attack",
    14: "Exploitation Attack"
}

# Apply mapping to create a new column
df["Generalized_Label"] = df["Label"].map(attack_mapping)

# Create a numeric version of Generalized_Label
df["Generalized_Label_Num"], _ = pd.factorize(df["Generalized_Label"])

# Display unique values
print(df[["Generalized_Label", "Generalized_Label_Num"]].drop_duplicates())
```

```python
from sklearn.model_selection import train_test_split

# Define features and labels
X = df.drop(columns=["Label"])  # Drop non-numeric labels
y = df["Label"]  # Use the generalized label

# Split into training and test sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
print("Class distribution in training set:\n", y_train.value_counts())
```

```python
from imblearn.under_sampling import RandomUnderSampler

# Define undersampling strategy (reduce BENIGN class)
undersampler = RandomUnderSampler(sampling_strategy={"BENIGN": 300000}, random_state=42)

# Apply undersampling
X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train, y_train)

print("New class distribution in training set:")
print(pd.Series(y_train_resampled).value_counts())
```

```python
# Combine X and y back into a DataFrame
df_resampled = pd.DataFrame(X_train_resampled, columns=X_train.columns)
df_resampled["Label"] = y_train_resampled  # Add the label column

# Save to CSV
df_resampled.to_csv("resampled_dataset.csv", index=False)

print("Resampled dataset saved as resampled_dataset.csv")
```
